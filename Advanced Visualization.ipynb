```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\tcaal\Desktop\Learning Python\Data\smoking.csv",
                encoding = 'ISO-8859-1')
```

```python
smokers = df[df['smoke'] == 'Yes']
```

```python
smokers
```

```python
men = smokers[smokers['gender'] == 'Male']
female = smokers[smokers['gender'] == 'Female']
```

```python
fig, axs = plt.subplots(1,2) # Similar to facet wrap

sns.histplot(female['age'], bins = 30, ax = axs[0])
axs[0].set_title('Age Distribution of Female Smokers')
sns.histplot(men['age'], bins = 30, ax = axs[1])
axs[1].set_title('Age Distribution of Male Smokers')

plt.tight_layout()
plt.show()
```

```python
from scipy import stats
```

```python
t_stat, p_value = stats.ttest_ind(men['age'],female['age'])
```

```python
print(t_stat)
print(p_value)
```

```python
house = pd.read_csv(r"C:\Users\tcaal\Desktop\Learning Python\Data\real_estate.csv",
                encoding = 'ISO-8859-1')

house
```

```python
numeric = house[['bed','bath','acre_lot','house_size','price']]
numeric
```

```python
print(numeric.corr(method = 'pearson'))
print(numeric.corr(method = 'spearman'))
print(numeric.corr(method = 'kendall'))
spear = numeric.corr(method = 'spearman')
```

```python
sns.heatmap(spear, annot = True, cmap = 'coolwarm')
plt.title('Spearman Correlation')
plt.show()
```
