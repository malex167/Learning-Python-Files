```python
h = pd.read_csv(r"C:\Users\tcaal\Desktop\Learning Python\Data\insurance.csv",
                 encoding = 'ISO-8859-1')
```

```python
h.head()
```

```python
h.info()
```

```python
regions = h.groupby('region')['charges'].mean()
```

```python
regions.head()
```

```python
sns.barplot(regions, x = 'region', y = 'charges', hue = 'region')
plt.xlabel('Region')
plt.ylabel('Average Yearly Charges')
plt.title('Average Charges Per Region')

plt.show()
```

```python
sns.histplot(h, x = 'charges', hue = 'smoker')
plt.xlabel('Yearly Cost')
plt.ylabel('Frequency of Individuals')
plt.title('Distribution of Charges for Smokers and Non Smokers')

plt.show()
```

```python
sns.histplot(h, x = 'bmi', y = 'charges', bins = 30)
plt.xlabel('BMI')
plt.ylabel('Charges Yearly')
plt.title('Charges to BMI')

plt.show()
```

```python
sns.scatterplot(h, x = 'age', y = 'charges', hue = 'bmi', size = 'bmi', sizes = (20,200), palette = 'coolwarm')
plt.xlabel('Age')
plt.ylabel('Yearly Charges')
plt.title('Yearly Charges by Age and BMI')

plt.show()
```
