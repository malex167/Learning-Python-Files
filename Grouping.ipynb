```python
stats = df.groupby('education').agg({'outstanding_balance':['median','mean','count','sum']})
```

```python
stats['avg'] = stats['outstanding_balance']['sum']/stats['outstanding_balance']['count']
```

```python
# Similar to SQL Windows function

df['average balance by job'] = df.groupby('job')['outstanding_balance'].transform('mean')
```

```python
df
```

```python
df['outstanding balance minus mean'] = df.groupby('job')['outstanding_balance'].transform(lambda x: x - x.mean())

# This lambda function will examine the 'x', in this case our variable outstanding balance, and then subtract the mean of x so we get a difference
```

```python
df
```

```python
# can filter too

df['average balance for having housing loan'] = df[df['housing_loan'] == 'yes'].groupby('job')['outstanding_balance'].transform('mean')
```
