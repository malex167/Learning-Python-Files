```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [5,10,15]
```

```python
plt.plot(x,y, 
         label = 'First Line',
         color = 'green',
         marker = '^',
         markersize = 10,
         linestyle = '--',
         markeredgecolor = 'blue')

z = [2,4,6]
plt.plot(x,z, 
         label = 'Second Line',
         color = 'red',
         marker = '^',
         markersize = 10,
         linestyle = '--',
         markeredgecolor = 'green')

plt.title('Matplotlib Title', fontdict = {'fontname': 'Times New Roman', 'fontsize': 20})
plt.xlabel('X Axis')
plt.ylabel('Y axis')

plt.xticks([1,2,3])
plt.yticks([0,5,10,15,20])
plt.legend()

plt.show()
```
