```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [5,10,15]
```

```python
plt.figure(figsize = (10,10))

plt.plot(x,y, 
         label = 'First Line',
         color = 'green',
         marker = '^',
         markersize = 10,
         linestyle = '--',
         markeredgecolor = 'blue')

z = [2,4,6]
plt.plot(x,z, 
         label = 'Second Line',
         color = 'red',
         marker = '.',
         markersize = 10,
         linestyle = '--',
         markeredgecolor = 'green')

plt.title('Matplotlib Title', fontdict = {'fontname': 'Times New Roman', 'fontsize': 20})
plt.xlabel('X Axis')
plt.ylabel('Y axis')

plt.xticks([1,2,3])
plt.yticks([0,5,10,15,20])
plt.legend()

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\tcaal\Desktop\Learning Python\Data\real_estate.csv",
                encoding = 'ISO-8859-1')
```

```python
df
```

```python
news = df[df['state'].isin([ 'New Jersey','New York'])]
```

```python
sns.barplot(news, x = 'state', y = 'price',
            estimator = 'median')
plt.plot()
```

```python
sns.set_style('dark')
```

```python
sns.barplot(news, x = 'state', y = 'price',
            estimator = 'mean')
plt.plot()
```

```python
filter = df[df['house_size'] < 10000]

sns.regplot(filter, x = 'house_size', y = 'price', line_kws = {'color':'red'})

plt.ticklabel_format(style = 'plain', axis = 'y')

plt.show()
```
